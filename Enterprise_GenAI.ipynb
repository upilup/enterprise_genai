### Question 1
**What are the emerging trends in GenAI adoption between 2022 and 2024, including training hours, new roles, and employee impact?**

We'll group by adoption year and GenAI tool with training hours, new roles, and total employees impacted columns. Then, use average of those columns to see the trends.

```python
year_mean = df.groupby(['adoption_year', 'genai_tool'])[['training_hours_provided', 'new_roles_created', 'total_employees_impacted']].mean().round(2)
year_mean
```

Then, let's breakdown with each GenAI tools in each columns of total training hours, new roles, and total employees impacted values based on years.

```python
summary_training = df.groupby(['adoption_year', 'genai_tool'])['training_hours_provided'].mean().round(2).reset_index()

# Add a new column with the rank
summary_training['rank'] = summary_training.groupby('adoption_year')['training_hours_provided'].rank(ascending=False)

# Sort for readability
training_ranked = summary_training.sort_values(['adoption_year', 'rank'])

# Display results
for year in [2022, 2023, 2024]:
    print(f"\nTraining Hours Ranking for {year}:")
    year_df = training_ranked[training_ranked['adoption_year'] == year]
    for idx, row in year_df.iterrows():
        print(f"  {int(row['rank'])}. {row['genai_tool']} — {row['training_hours_provided']:,.2f} hours")
print()
for year in [2022,2023,2024]:
    subset = summary_training[summary_training['adoption_year'] == year]
    min_row = subset.loc[subset['training_hours_provided'].idxmin()]
    max_row = subset.loc[subset['training_hours_provided'].idxmax()]
    print(f"In {year}, GenAI tool '{max_row['genai_tool']}' had the highest training hours provided: {max_row['training_hours_provided']:.2f} hours")
    print(f"In {year}, GenAI tool '{min_row['genai_tool']}' had the lowest training hours provided: {min_row['training_hours_provided']:.2f} hours")
    print()
```

```python
summary_roles = df.groupby(['adoption_year', 'genai_tool'])['new_roles_created'].mean().reset_index()

# Add a new column with the rank
summary_roles['rank'] = summary_roles.groupby('adoption_year')['new_roles_created'].rank(ascending=False)

# Sort for readability
roles_ranked = summary_roles.sort_values(['adoption_year', 'rank'])

# Display results
for year in [2022, 2023, 2024]:
    print(f"\nNew Roles Created Ranking for {year}:")
    year_df = roles_ranked[roles_ranked['adoption_year'] == year]
    for idx, row in year_df.iterrows():
        print(f"  {int(row['rank'])}. {row['genai_tool']} — {row['new_roles_created']:,.2f} new roles created")
print()
for year in [2022,2023,2024]:
    subset = summary_roles[summary_roles['adoption_year'] == year]
    min_row = subset.loc[subset['new_roles_created'].idxmin()]
    max_row = subset.loc[subset['new_roles_created'].idxmax()]
    print(f"In {year}, the highest new roles created by GenAI tool '{max_row['genai_tool']}': {max_row['new_roles_created']:.2f} roles")
    print(f"In {year}, the lowest new roles created by GenAI tool '{min_row['genai_tool']}': {min_row['new_roles_created']:.2f} roles")
    print()
```

Purpose Use Tukey's Rule: 
1. Identify outliers using a robust method (not affected by skewed data like Z-score might be).
2. Tukey uses interquartile range (IQR) to flag values that are abnormally high or low.
3. Especially important in real-world HR or organizational datasets where big outliers could distort trends.

```python
def tukey_flags(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

df['outlier_training'] = tukey_flags(df['training_hours_provided'])
df['outlier_roles'] = tukey_flags(df['new_roles_created'])
df['outlier_impact'] = tukey_flags(df['total_employees_impacted'])

# Outlier count summary
print("Tukey Outlier Counts:")
print(df[['outlier_training', 'outlier_roles', 'outlier_impact']].sum())
```

Growth Rate Calculation<br>
Purpose: 
1. Measure the relative growth of employee impact per GenAI tool.
2. Normalizes differences between tools by comparing how much they’ve changed over time.
3. Gives a simple yet powerful metric to compare impact trajectories.

```python
# Growth Rate Calculation (Relative % change from 2022 to 2024)
growth_df = summary_employees.pivot(index='genai_tool', columns='adoption_year', values='total_employees_impacted')
growth_df['pct_change_2022_2024'] = ((growth_df[2024] - growth_df[2022]) / growth_df[2022] * 100).round(2)
print(growth_df[['pct_change_2022_2024']].sort_values(by='pct_change_2022_2024', ascending=False))
```

Insight: Only LLaMA shows a clearly increasing trend (+1.16%), while Claude (-0.78%) and Gemini (-0.50%) are declining in employee impact.

Trendline Slope via Linear Regression<br>
Purpose: 
1. Quantify the direction and strength of trend over time.
2. The slope tells you whether a tool’s employee impact is rising, flat, or falling.
3. Useful for building Tableau line plots that aren’t just visual — they’re statistically backed.

```python
# Trendline Slope per Tool (Linear Regression)
from scipy.stats import linregress
slopes = {}
for tool in summary_employees['genai_tool'].unique():
    subset = summary_employees[summary_employees['genai_tool'] == tool]
    slope, _, _, _, _ = linregress(subset['adoption_year'], subset['total_employees_impacted'])
    slopes[tool] = round(slope, 2)
print(sorted(slopes.items(), key=lambda x: x[1], reverse=True))
```

Insight: LLaMA has the steepest upward trend. Claude the steepest decline.

Between 2022–2024:

- LLaMA consistently grew in impact, showing the strongest upward trend (+1.16%, slope +58.5) and may indicate increasing integration into broader workforce roles.
- Claude experienced a decline both in total impact and stability, signaling possible deprioritization or saturation in its use.
- Gemini showed a moderate downtrend, while Groq and ChatGPT remained stable.
- Mixtral maintained low impact but slightly increased.

### Question 9
**Is there a pattern in how GenAI adoption affected productivity, depending on the tool used?**

H₀ (Null Hypothesis):
The mean productivity change is the same across all GenAI tools.

H₁ (Alternative Hypothesis):
At least one GenAI tool has a different mean productivity change.

```python
# Group by GenAI tool and year to calculate average productivity change
avg_prod = df.groupby('genai_tool')['productivity_change'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=avg_prod, x='genai_tool', y='productivity_change', palette='Set2')

# Adding bar values in the center
for i, row in avg_prod.iterrows():
    plt.text(i, row['productivity_change'] / 2, f"{row['productivity_change']:.1f}%", 
             ha='center', va='center', color='black', fontweight='bold')

# Titles and labels
plt.title("Average Productivity Change by GenAI Tool (2022–2024)", fontsize=16)
plt.ylabel("Avg. Productivity Change (%)")
plt.xlabel("GenAI Tool")
plt.tight_layout()
plt.show()
```

```python
df['productivity_change'].describe()
```

```python
# 2. ANOVA: Test whether mean productivity change differs by GenAI tool
from scipy.stats import f_oneway

# Prepare data for ANOVA
tool_groups = [group['productivity_change'].values for name, group in df.groupby('genai_tool')]

# Perform ANOVA
anova_result = f_oneway(*tool_groups)

print("ANOVA Results: Is productivity change significantly different across tools?")
print(f"F-statistic: {anova_result.statistic:.4f}")
print(f"P-value: {anova_result.pvalue:.4f}")
```

Since the p-value (0.1590) > 0.05, we fail to reject the null hypothesis.

Interpretation:
There is no statistically significant evidence that different GenAI tools lead to different average productivity changes.

From the data, we see that Gemini and LLaMA led the way in average productivity improvement across all tools, followed closely by Groq and ChatGPT. While the differences between tools weren’t massive, the consistent performance of these top tools suggests that organizations using them might have seen slightly better efficiency gains. The results also highlight how most tools offer relatively similar outcomes in productivity, which could mean that success depends just as much on how the tools are used as on the tools themselves.

### Question 10
**What projections can be made about productivity trends in 2025 based on GenAI adoption between 2022–2024?**

```python
from sklearn.linear_model import LinearRegression

# Step 1: Prepare the Data
avg_productivity = df.groupby('adoption_year')['productivity_change'].mean().reset_index()
X = avg_productivity['adoption_year'].values.reshape(-1, 1)  # Feature
y = avg_productivity['productivity_change'].values           # Target

# Step 2: Train Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# Step 3: Predict for 2025
future_year = np.array([[2025]])
future_pred = model.predict(future_year)[0]

# Step 4: Plot the trend
plt.figure(figsize=(8,5))
sns.lineplot(data=avg_productivity, x='adoption_year', y='productivity_change', marker='o', label='Actual')
plt.scatter(2025, future_pred, color='red', label=f'2025 Projection: {future_pred:.2f}%')
plt.title("Projected Productivity Change in 2025 Based on GenAI Adoption")
plt.xlabel("Year")
plt.ylabel("Average Productivity Change (%)")
plt.xticks([2022, 2023, 2024, 2025])
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Between 2022 and 2024, productivity stayed pretty consistent — it was 18.52% in 2022, dropped a bit in 2023 to around 18.42–18.44%, then slightly went up in 2024 to about 18.44–18.46%. But based on the linear regression, the prediction for 2025 is 18.39%, which shows a small decline.

This might mean that even though GenAI tools are still being adopted, they’re not pushing productivity as much as before. So, companies may need to think about how they’re actually using these tools if they want to keep seeing strong results.
